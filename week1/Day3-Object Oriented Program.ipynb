# init method
*Special Method to define variables in class
*Same like constructor
*Called automatically

```python
class Training:
  def __init__(self,name,fee,duration):
    print("This is init function")
    self.n = name
    self.f = fee
    self.t = duration

  def subjects(self):
    print("Course name - {}, Fee - {} , Duration - {}".format(self.n,self.f,self.t))
t1 = Training("Python",5000,"7 days")
t2 = Training("Ml",7000,"5 days")
t1.subjects()
t2.subjects()
```

Inheritance
* Class using the property/method of another class
1.Single Level Inheritance
2.Multi Level Inheritance
3.Multiple Inheritance

```python
# Single Level Inheritance
```

```python
class A:
  def method1(self):
    print("This is Method 1")
class B(A):
  def method2(self):
    print("This is Method 2")
A1 = A()
B1 = B()
B1.method1()
A1.method2()
```

```python
# Multi Level Inheritance
```

```python
class A:
  def method1(self):
    print("This is Method 1")
class B(A):
  def method2(self):
    print("This is Method 2")

class C(B):
  def method3(self):
    print("This is Method 3")
A1 = A()
B1 = B()
C1 = C()
A1.method1()
B1.method1()
B1.method2()
C1.method1()
C1.method2()
C1.method3()
```

```python
# Multiple Inheritance
```

```python
class A:
  def method1(self):
    print("This is Method 1")
class B:
  def method2(self):
    print("This is Method 2")

class C(A,B):
  def method3(self):
    print("This is Method 3")

A1 = A()
B1 = B()
C1 = C()
A1.method1()
B1.method2()
C1.method1()
C1.method2()
C1.method3()
```

```python
# Method Overriding
class A:
  def __init__(self):
    print("This is init of A")
  def info(self):
    print("This is info of A")
  def method1(self):
    print("This is Method 1")
class B(A):
  def __init__(self):
    print("This is init of B")
  def info(self):
    print("This is info of B")
  def method2(self):
    print("This is Method 2")
A1 = A()
B1 = B()
A1.info()
B1.info()
```

```python
# Method Overriding | Polymorphism
class A:
  def __init__(self):
    print("This is init of A")
  def method1(self):
    print("This is Method 1")
class B():
  def __init__(self):
    print("This is init of B")
  def method2(self):
    print("This is Method 2")
class C(B,A):
  def __init__(self):
    super().__init__()
    print("This is init of C")
  def method3(self):
    print("This is Method 3")
```

```python
C1 = C()
```

```python
# Sub class - Class inside Class
```

```python
class Student:
  def __init__(self,name,roll,brand,cpu):
    self.name = name
    self.roll = roll
    self.b = brand
    self.c = cpu
    self.lap = self.Laptop(self.b,self.c)
  def show(self):
    print(self.name,self.roll)
    self.lap.show()
  class Laptop:
    def __init__(self,brand,cpu):
      self.brand = brand
      self.cpu = cpu
    def show(self):
      print(self.brand,self.cpu)
```

```python
s1 = Student("Arnab",1,"Dell","i7")
s2 = Student("Bob",2,"Asus","i9")
s1.show()
s2.show()
```
