```python
i = 0
while i<10:
  print("Value of i is {}".format(i))
  i = i + 1
```

```python
#For Loop
```

```python
for i in range(10):
  print("The value of i is {}".format(i))
```

```python
for i in range(10,20):
  print("The value of i is {}".format(i))
```

```python
for i in range(10,20,3):
  print("The value of i is {}".format(i))
```

```python
fruits = ["Apple","Mango","Cherry","Banana","Strawberry"]
for i in range(len(fruits)):
  print(fruits[i])
```

```python
#foreach loop
fruits=["Apple","Mango","Cherry","Banana","Strawberry"]
for i in fruits:
  print(i)
```

```python
#Break and continue statement
```

```python
for i in range(10):
  if(i == 5):
    break
  print("The value of i is {}".format(i))
```

```python
for i in range(10):
  if(i == 5):
    continue
  print("The value of i is {}".format(i))
```

```python
for i in range(10):
  print("The value of i is {}".format(i))
  if(i == 5):
    continue
  print("Umang")
```

```python
# 1. Function with argument
```

```python
def greet_user(name):
  print("Hello,how are you {}".format(name))
```

```python
greet_user()
```

```python
greet_user("Umang")
```

```python
def greet_user_city(name,city):
  print("Hello,how are you {}? How is the weather in {}".format(name,city))
```

```python
greet_user_city("Umang","Mumbai")
```

```python
greet_user_city("Mumbai","Umang")
```

```python
greet_user_city(city="Mumbai",name="Umang")
```

```python
#3.Function with default/optional argument
```

```python
def greet_user_city(city,name = "Unknown"):
  print("Hello,how are you {}? How is the weather in {}".format(name,city))
```

```python
greet_user_city(city="Mumbai",name="Umang")
```

```python
greet_user_city(city="Mumbai")
```

```python
# 4. Function with return statement
```

```python
def greet_user_city_return(name,city):
  print("Hello,how are you {}? How is the weather in {}".format(name,city))
```

```python
x = greet_user_city_return("Umang","Mumbai")
```

```python
print(x)
```

```python
def greet_user_city_return(name,city):
  return("Hello,how are you {}? How is the weather in {}".format(name,city))
```

```python
x = greet_user_city_return("Umang","Mumbai")
```

```python
print(x)
```
