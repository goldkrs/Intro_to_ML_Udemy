```python
# Core Datatype
```

List
* Group of heterogenous data(different data type)
* It is identified as ['a',7.2,'c',1,'e']
* It is mutable(can be modified)

Tuple
* Group of heterogenous data(different data type)
* It is identified as ('a',7.2,'c',1,'e')
* It is immutable(cannot be modified)

Dictionary
* Stores data in the form of key-value pairs
* It is identified as {"key1":"value1","key2":"value2"}
* It is mutable(can be modified)

Array
* Group of homogenous data(similar data type)
* It is identified as [2 3 4 5 6]
* It can be multidimensional
* It is immutable(cannot be modified)

Set
* Unordered collection of unique data
* It is identified as {a,b,c,d,e}
* It is mutable(can be modified)
