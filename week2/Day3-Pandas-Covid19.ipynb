```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/covid_19_data.csv")
data.head()
```

```python
data = data.drop(["SNo","Last Update"],axis = 1)
data.head()
```

```python
data.info()
```

```python
data["Confirmed"] = data["Confirmed"].astype("int64")
data["Deaths"] = data["Deaths"].astype("int64")
data["Recovered"] = data["Recovered"].astype("int64")
data["ObservationDate"] = pd.to_datetime(data["ObservationDate"])
```

```python
data.info()
```

```python
data["Active"] = data["Confirmed"] - data["Deaths"] - data["Recovered"]
data.head()
```

```python
filter = data['ObservationDate'] == data['ObservationDate'].max()
latest_data = data[filter]
latest_data.head()
```

```python
latest_data.shape
```

```python
latest_data["Country/Region"].value_counts()
```

```python
countries = latest_data.groupby("Country/Region")[["Confirmed","Deaths","Recovered","Active"]].sum()
countries = countries.reset_index()
countries
```

```python
filter1 = countries["Country/Region"] == 'India'
filter2 = countries["Country/Region"] == 'Mainland China'
countries[filter1 | filter2]
```

```python
# Top 10 countries having maxm no of active cases
```

```python
countries.sort_values("Active",ascending = False).head(10)
```

```python
### total case for each date
date_wise_case = data.groupby("ObservationDate")[["Confirmed","Deaths","Recovered","Active"]].sum().reset_index()
date_wise_case
```
