```python
import numpy as np
```

```python
### Create an array
```

```python
a1 = np.array([12,31,4,2,2,345,1,3])  # 1d array
a2 = np.array([[12,31,4,2,2,345,1,3]])  # 2d array
a3 = np.array([[12,31,4,2,2,345,1,3],[12,31,4,2,2,345,1,3]])  # 2d array
a4 = np.array([[[12,31,4,2,2,345,1,3],[12,31,4,2,2,345,1,3]]])  # 3d array
a5 = np.array([[[12,31,4,2,2,345,1,3],[12,31,4,2,2,345,1,3]],
               [[12,31,4,2,2,345,1,3],[12,31,4,2,2,345,1,3]]])  # 3d array
```

```python
print(len(a1))
print(len(a2))
print(len(a3))
print(len(a4))
print(len(a5))
```

```python
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
print(a4.ndim)
print(a5.ndim)
```

```python
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
print(a4.dtype)
print(a5.dtype)
```

```python
print(a1.shape)
print(a2.shape)
print(a3.shape)
print(a4.shape)
print(a5.shape)
```

```python
print(np.max(r1))
print(np.max(r1,axis=0)) #row
print(np.max(r1,axis=1)) #column
```

```python
print(np.min(r1))
print(np.min(r1,axis=0)) #row
print(np.min(r1,axis=1)) #column
```

```python
print(np.mean(r1))
print(np.mean(r1,axis=0)) #row
print(np.mean(r1,axis=1)) #column
```

```python
print(np.sum(r1))
print(np.sum(r1,axis=0)) #row
print(np.sum(r1,axis=1)) #column
```

```python
print(np.argmax(r1))
print(np.argmax(r1,axis=0)) #row
print(np.argmax(r1,axis=1)) #column
```

```python
print(np.argmin(r1))
print(np.argmin(r1,axis=0)) #row
print(np.argmin(r1,axis=1)) #column
```
