```python
def rating_label(x):
  if x <= 2.5 :
    return "Bad"
  elif x <= 3.5:
    return "Good"
  else:
    return "Excellent"
```

```python
# Join = Inner | Left | Right | Full Join
```

```python
d1 = {"UserId":[1,2,3,4,5],"Name":["Umang","Rahul","Rohit","John","Kasper"]}
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {"UserId":[1,2,3,4,6],"City":["Delhi","Mumbai","Delhi","Chennai","Kolkata"]}
df2 = pd.DataFrame(d2)
df2
```

```python
# Inner Join
df = pd.merge(df1,df2,on="UserId",how="inner")
df
```

```python
# Left Join
df = pd.merge(df1,df2,on="UserId",how="left")
df
```

```python
# Right Join
df = pd.merge(df1,df2,on="UserId",how="right")
df
```

```python
# Full Join
df = pd.merge(df1,df2,on="UserId",how="outer")
df
```
