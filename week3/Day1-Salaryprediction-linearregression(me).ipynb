```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Salary_Data.csv")
data.head()
```

```python
data.info()
```

```python
x = data[["YearsExperience"]].values # input/feature
x
```

```python
y = data[["Salary"]].values # target/output
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.8,random_state=0)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
xtest
```

```python
ytest
```

```python
ypred = model.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(ytest,ypred)
r2
```

```python
### Line of regression
plt.scatter(xtrain,ytrain,label="Actual Samples",s=120,color="blue")
plt.scatter(xtrain,model.predict(xtrain),label="Predicted Samples",s=120,color="red")
plt.plot(xtrain,model.predict(xtrain),label="line of regression",linestyle="dotted",color="purple")
```

```python
### Line of regression(Testing Samples)
plt.scatter(xtest,ytest,label="Actual Samples",s=120,color="blue")
plt.scatter(xtest,ypred,label="Predicted Samples",s=120,color="red")
plt.plot(xtest,ypred,label="line of regression",linestyle="dotted",color="purple")
```

```python
m = model.coef_
c = model.intercept_
print(m,c)
```
