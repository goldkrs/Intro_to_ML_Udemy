```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/income.csv")
data.head()
```

```python
plt.scatter(data['Age'],data['Income($)'])
```

```python
x = data[['Age','Income($)']]
x.columns = ['Age','Income']
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
x.columns = ['Age','Income']
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
y = km.fit_predict(x)
```

```python
y
```

```python
x['Cluster'] = y
```

```python
x
```

```python
x.head(10)
```

```python
centroids = km.cluster_centers_
centroids
```

```python
x1 = x[x['Cluster']==0]
x2 = x[x['Cluster']==1]
x3 = x[x['Cluster']==2]
plt.scatter(x1['Age'],x1['Income'],color='blue',label='Cluster 1',s = 100)
plt.scatter(x2['Age'],x2['Income'],color='red',label='Cluster 2',s = 100)
plt.scatter(x3['Age'],x3['Income'],color='purple',label='Cluster 3',s = 100)
plt.scatter(centroids[:,0],centroids[:,1],color='orange',marker = "*",label="Centroids",s=150);
```
